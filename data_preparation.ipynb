```python
# Removal of upc values from the sales_ab dataset that do not ever appear in the out_of_stock data.
''' Justification for removing the upc values that do not ever appear in the out_stock_data:
      Over the time period considered, these codes have never appeared in out_of_stock data so from
      a modelling perspective there is no indication they have ever been 'out_of_stock' hence they will
      be assigned a very low (or zero) probability of being recorded as out of stock.
      For this reason upcs that do not appear in the out_of_stock file are removed.
'''
A = set(sales_data['upc'])
B = set(out_of_stock['upc'])
a = list(B.intersection(A)) # upcs common to both A and B
len(a)
# Note we choose upcs that also occur in the sales data set( otherwise we have an absence of sales data for some upcs)

s = list(sales_data['upc'])
s_bool = [x in a  for x in s ]

sales_data = sales_data[s_bool]   # removal of never occuring upc values

```

```python
'''
1)
    a) Note the variable 'scanned_lbs' is entirely 0 and can be deleted from the sales_ab data



    b) Cases where the upc description takes on different forms

                upc        upc_description
24990    1820097009    BUD LIGHT CALNR 12CT
46308    1820097009    BUD LIGHT  CALNR 12CT
2570445  1820097009    BUD LIGHTCALNR 12CT




2) Rationale for left join:
   'storenums' that do not ever appear in the sales data (but in the out_stock data) will not have any recorded sales information -- from a
   modelling perspective this is a lack of information which results in greater uncertainty in predictions for these stores (the financial data will have NaNs)
   To avoid this situation, we settle for a left join, i.e., all the required elements from the sales_data are kept and 'storenums' that do not have sales data are dropped
'''

new_merge = sales_data[['upc', 'storenum',
       'scanned_retail_dollars', 'scanned_movement',
       'gross_margin_dollars', 'division_code', 'manager_on_floor',
       'new_year_mth_wk']].merge(out_of_stock[['upc', 'upc_description', 'date_key', 'storenum', 'new_year_mth_wk', 'dsd_vendor_name']]
                                 , how = 'left', suffixes = ('_x', None)
                                 , on = ['new_year_mth_wk', 'upc',  'storenum']

                                )



```

```python
# Creation of a modelling dataset


''' We need to aggregate over the 'upc', 'upc_description' , 'storenum' and 'new_year_mth_wk'  codes, 'manager_on_floor' and 'dsd_vendor_name'
in order to obtain the number of out_of_stocks per week

Note : That there is a 1-1 correspondence between 'upc' and 'upc_description', so it is redundant to include both
in a grouping function. We do so nevertheless - for later illustrations.
'''


x = ["target_out_stock", "gross_margin_dollars", "scanned_retail_dollars", 'scanned_movement']
df = new_merge.groupby(['upc', 'storenum', 'new_year_mth_wk', 'upc_description'], axis = 0)[x].sum()

new_merge_dat = df.reset_index()
```

```python
#Constructing a 'Wks' and 'year' column from the 'new_year_mth_wk' variable
# Easier way: use str.slice for a pandas dataframe.
a = list(new_merge_dat['new_year_mth_wk'])
b  = [x.split() for x in a]
wks = [ ''.join([l[3], l[4]]) for l in b]
year = [l[0] for l in b]

new_merge_dat['wks'] = wks
new_merge_dat['year'] = new_merge_dat['new_year_mth_wk'].str.slice(start = 0, stop = 5)
```

```python
new_merge_dat.head()
```

```python
# Checking for negative values in the scanned_retail_dollars column


# We remove this data where it does not affect out_of_stock values --- rationale : these are returns that have not impacted out-of-stock shelves;  we will regard 'returns' as abnormal
# behaviour but to be tolerated.  A negative profit (as indicated by the gross_margin_dollars var.) may be indicative of other factors -- discounts, sales, special events , etc

a = new_merge_dat['scanned_retail_dollars'] < 0
b = new_merge_dat['target_out_stock'] > 0

sum(a&b)
```

```python
# So in this case sum(a&b) equals 1 , that is  1 row is deleted.
u = [not(x) for x in a & b]
new_merge_dat = new_merge_dat[u]
```

```python
''' We initially dropped the following variables :
  'scanned_lbs'
  'division_code'; has a single value, namely 705


  The variables below are constructed in order to be used later in an optimization formulation
  Derived variables (features)
  1) profit_unit = 'gross_margin_dollars' / 'scanned_movement'

  2) retail_price_unit = 'scanned_retail_dollars' / 'scanned_movement'


'''



new_merge_dat = new_merge_dat.assign(profit_unit = new_merge_dat.gross_margin_dollars/new_merge_dat.scanned_movement)

new_merge_dat = new_merge_dat.assign(retail_price_unit  = new_merge_dat.scanned_retail_dollars/new_merge_dat.scanned_movement)
```

```python
'''
    Removing the upc_description' , 'storenum' combination for which there is no indication of out of stock over the modelling period;
    that is products in these stores were never out of stock over the modelling period.

'''
def transform(dat):
    # Creation of features (derived variables) from the initial modelling dataset
    dat['scanned_retail_dollars:scanned_movement'] = dat['scanned_retail_dollars']*dat['scanned_movement']

    # binomial target: (0, 1) vector
    dat['logistic_target'] = [1 if x >= 1 else 0 for x in dat['target_out_stock']]



    '''
    Removing 'upc_description' , 'storenum' combination for which there is no indication of out of stock over the modelling period; that is products in these stores
    were never out of stock over the modelling period.

    '''
    out_stock_cnts = dat.groupby(['upc_description', 'storenum'])['target_out_stock'].sum('target_out_stock')
    out_stock_cnts = out_stock_cnts.reset_index()
    y = out_stock_cnts['target_out_stock'] >= 1
    dat = dat.merge(out_stock_cnts[['upc_description', 'storenum']][y], on = ['upc_description', 'storenum'], how = 'inner', suffixes = ("_x", None))
    return(dat)

```

```python
# Applying the transform function

new_merge2 = transform(dat = new_merge_dat)
new_merge2.shape
```
